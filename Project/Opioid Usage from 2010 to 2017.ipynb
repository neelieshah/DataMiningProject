```python
df17 = pd.read_csv('Opioid2017.csv')
population = pd.read_csv('population.csv')

df17 = df17.rename(columns = {"Natural and Semisynthetic Opioids (e.g. oxycodone, hydrocodone)" : "Semisynthetic", 
                       "Synthetic Opioids, other than Methadone (e.g. fentanyl, tramadol)" : "Synthetic"})
df17['Year'] = 2017
df17['Total Population'] = population['2017'].astype(float)
df17 = df17.replace("NSD",0)
df17.Synthetic = df17.Synthetic.astype(float)
df17.Semisynthetic = df17.Semisynthetic.astype(float)
df17.Heroin = df17.Heroin.astype(float)
df17.Methadone = df17.Methadone.astype(float)
df17['Year']=df17.Year.astype(float)
```

```python
df10 = pd.read_csv('Opioid2010.csv')
df10 = df10.rename(columns = {"Natural and Semisynthetic Opioids (e.g. oxycodone, hydrocodone)" : "Semisynthetic", 
                       "Synthetic Opioids, other than Methadone (e.g. fentanyl, tramadol)" : "Synthetic"})
df10['Year'] = 2010
df10.Year.astype('int32')
df10['Total Population'] = population['2010'].astype(float)

df10 = df10.replace("NSD",0)
df10.Synthetic = df10.Synthetic.astype(float)
df10.Semisynthetic = df10.Semisynthetic.astype(float)
df10.Heroin = df10.Heroin.astype(float)
df10.Methadone = df10.Methadone.astype(float)
df10['Year']=df10.Year.astype(float)
```

```python
dfMega = pd.concat([df17,df16,df15,df14,df13,df12,df11,df10])
```

```python
dfMega['Total Deaths'] = dfMega.Semisynthetic + dfMega.Synthetic + dfMega.Methadone + dfMega.Heroin
dfMega['Semisynthetic %'] = dfMega.Semisynthetic/dfMega['Total Deaths'] 
dfMega['Synthetic %'] = dfMega.Synthetic/dfMega['Total Deaths'] 
dfMega['Methadone %'] = dfMega.Methadone/dfMega['Total Deaths'] 
dfMega['Heroin %'] = dfMega.Heroin/dfMega['Total Deaths'] 
dfMega['Deaths %'] = dfMega['Total Deaths']/dfMega['Total Population']
```
